```python
#Task 1

data = pd.read_csv('/content/sample_data/visit_log.csv',sep=';')
data.head()
```

```python
def source_class(data):
    if data.traffic_source == 'yandex' or data.traffic_source == 'google':
        return 'organic'
    if data.traffic_source == 'paid' or data.traffic_source =='email' and data.region == 'Russia':
        return 'ad'
    if data.traffic_source == 'paid'or data.traffic_source =='email' and data.region != 'Russia':
        return 'other'
    return data.traffic_source
```

```python
data['source_type'] = data.apply(source_class, axis =1)
```
