```python
import pandas as pd
```

```python
#Task 1

data = pd.read_csv('/content/sample_data/ratings.csv')
data.head()
```

```python
def classification(row):
        if row['rating'] <= 2:
            return 'low'
        if 4 >= row['rating'] > 2:
            return 'average'
        if row['rating'] > 4:
            return 'high'
```

```python
data['class'] = data.apply(classification, axis=1)
data.head()
```

```python
#Task 2

data_reg = pd.read_csv('/content/sample_data/keywords.csv')
data_reg.head()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def region(row):
    if 'москва' in row['keyword'].lower() or 'тула' in row['keyword'].lower() or 'ярославль' in row['keyword'].lower():
        return 'Центр'
    if 'петербург'in row['keyword'].lower() or 'псков' in row['keyword'].lower() or 'мурманск' in row['keyword'].lower():
        return 'Северо-Запад'
    if 'владивосток' in row['keyword'].lower() or 'сахалин' in row['keyword'].lower() or 'хабаровск' in row['keyword'].lower():
        return 'Дальний Восток'
    return 'undefined'
```

```python
data_reg['region'] = data_reg.apply(region, axis=1)
data_reg[data_reg['region'] == 'Центр']
```
